```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/raw/youtube_shorts_tiktok_trends_2025.csv")

print(df.shape)
print(df.columns)
```

```python
print(df["views"].describe())
print(df["engagement_rate"].describe())
```

```python
plt.hist(df["views"], bins=50)
plt.title("Views distribution")
plt.show()
```

```python
print(df["views"].quantile([0.5,0.75,0.9,0.95,0.99]))
print(df["engagement_rate"].quantile([0.5,0.75,0.9,0.95]))
```

```python
viral = (df["views"] >= 100000) & (df["engagement_rate"] >= 0.06)

print(viral.mean())
```
